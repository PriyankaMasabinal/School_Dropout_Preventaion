```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# 1 Load dataset
df = pd.read_csv("CHINA_10k.csv")
```

```python
# 2 Check dataset shape
print("Dataset shape:", df.shape)
```

```python
# 3 View first rows
print(df.head())
```

```python
# 4 Check dataset info
print(df.info())
```

```python
# 5 Check missing values
print(df.isnull().sum())
```

```python
# 6 Remove unnecessary column
df.drop("student_id", axis=1, inplace=True)
```

```python
# 7 Remove duplicate rows
df.drop_duplicates(inplace=True)
```

```python
# 8 Handle missing values (numerical columns)
numeric_cols = df.select_dtypes(include=['int64','float64']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
```

```python
# 9 Handle missing values (categorical columns)
categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
# 10 Encode categorical variables
encoder = LabelEncoder()

for col in categorical_cols:
    df[col] = encoder.fit_transform(df[col])
```

```python
# 11 Verify cleaned dataset
print(df.head())
```

```python
# 12 Save cleaned dataset
df.to_csv("cleaned_student_dataset.csv", index=False)
```
